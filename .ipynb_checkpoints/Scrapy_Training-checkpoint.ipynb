# Prepare link_list

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import logging
import numpy as np
import re
import tqdm
import threading
import queue
```

```python
link_list=[]
for i in tqdm.tqdm(range(200000,337000)):
    link_list.append('http://www.newsmth.net/nForum/article/Intern/'+str(i)+'?ajax' )
```

# Process extends Thread Class

```python
class ShuimuSpider(Thread):
    def __init__(self,url,q):
        super(ShuimuSpider,self).__init__()
        self.url=url
        self.q=q
        
    def run(self):
        self.parse_page()
        
    def send_request(self,url):
        try:
            html = requests.get(url,timeout=2)
        except Exception as e:
            print("fail at %s" % url)
        else:
            return html
    def parse_page(self):
        response = self.send_request(self.url)
        if response.status_code==200:
            response_bs=response.content
            try:
                bs = BeautifulSoup(response_bs,'html.parser')


                text=''
                store_data={
                    'time':[],
                    'title':[],
                    'detail':[]
                }
                text = bs.p.text
                # .p 对应html中的<p>, 观察帖子不难发现，几乎所有的招聘正文都在帖子的<p>内
                # .text 对应输出文本

                #分别通过特殊字符串的位置给标题、时间、正文定位
                #position for title:
                Title_Position1=text.find('标  题: ')+len('标  题: ')
                Title_Position2=text.find('  发信站: 水木社区')
                Title = text[Title_Position1:Title_Position2]
                #text[x:y]进行切片

                #position for time:
                Time_Position1=text.find('  发信站: 水木社区')+len('  发信站: 水木社区')
                Time_Position2=text.find('), 站内')+1
                Time=text[Time_Position1:Time_Position2]

                #position for detail
                Detail_Position1=text.find('), 站内')+len('), 站内')+4
                Detail_Position2=text.find('※ 来源:·水木社区')
                Detail=text[Detail_Position1:Detail_Position2]

                store_data['time'].append(Time)
                store_data['title'].append(Title)
                store_data['detail'].append(Detail)
                #字典赋值
                #运行成功，index序列加一

                self.q.put(store_data)
            except:
                return
        else:
            return
            
```

# RUN IT!

每次爬取数目超过150时会出现中断，则分段进行

```python
def Fast_Scrapy(x,y,store_data):
    start = time.time()
    q= queue.Queue(0)

    Thread_list=[]
    for url in link_list[x:y]:
        p = ShuimuSpider(url,q)
        p.start()
        Thread_list.append(p)

    for i in Thread_list:
        i.join()

    while not q.empty():
        data = q.get()

        store_data['time'].append(data['time'])
        store_data['title'].append(data['title'])
        store_data['detail'].append(data['detail'])
    end = time.time()
    print('Process + Queue多线程爬虫的总时间为：', end-start)
    print('Main process Ended!')
```

```python
store_data={
    'time':[],
    'title':[],
    'detail':[]
}

period = np.arange(0,137000,150)

for i in tqdm.tqdm(period):
    Fast_Scrapy(i,i+150,store_data)
```

## 将岗位职责及以后的内容填写进'requirement'

```python
x= re.search('(职责|内容|要求)(.|\n)*',df['detail'][6]).span()
```

```python
req = re.compile('(职责|内容|要求)(.|\n)*')
```

```python
df['requirement']=np.zeros(len(df))
```

```python
def find_req(x):
    try: 
        re.search(req,x).span()
        return x[re.search(req,x).span()[0] : re.search(req,x).span()[1]]
    except:
        return ''
```

```python
df['requirement']=df['detail'].apply(find_req)
```

```python
df = df[df['requirement']!='']
```

# Cluster [KMeans]

```python
from sklearn.feature_extraction.text import TfidfVectorizer,TfidfTransformer
from sklearn.cluster import KMeans
import jieba
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
```

```python
import jieba_fast
```

## Load StopWords

```python
stopwords = pd.read_csv('stopwords/baidu_stopwords.txt',
                        quoting=3,sep='\t',names=['stopword'],encoding='utf-8')
stopwords = stopwords['stopword'].values
```

## shuffle the dataset

```python
from sklearn.utils import shuffle

df_shuffle = shuffle(df,random_state=42)
df_shuffle.reset_index(drop=True,inplace=True)
```

```python
text_train =[]
for i in tqdm.tqdm(range(len(df_shuffle))):
    text_train.append(str(df_shuffle['title_no_bracket'][i]+df_shuffle['requirement'][i]))
```

## Word Cut

```python
def preprocess_text(content_lines,sentences):
    for line in tqdm.tqdm(content_lines):
        try:
            segs=jieba_fast.cut(line,cut_all=False)
            segs = list(filter(lambda x:x.strip(),segs))
            segs=list(filter(lambda x:((x not in stopwords) and x != 'xa0' ),segs))
            sentences.append(str([x for x in segs]))
        except Exception:
            continue
```

```python
%%time
sentences = []
preprocess_text(text_train,sentences)
```

```python
pd.DataFrame(sentences).to_csv('sentence_jieba_10000_shuffle.csv')
```

```python
vectorizer = TfidfVectorizer()
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(vectorizer.fit_transform(sentences))
```

```python
tfidf.shape
```

## Reduce Dimensionality

```python
%%time
from sklearn.decomposition import KernelPCA

rbf_pca = KernelPCA(n_components=500,kernel='rbf')
tfidf_reduced_500 = rbf_pca.fit_transform(tfidf)
```

```python
rbf_pca_2 = KernelPCA(n_components=2,kernel='rbf')
tfidf_reduced_20 = rbf_pca_2.fit_transform(tfidf)
```

## KMeans

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import silhouette_score
```

```python
%%time

score=[]
for i in tqdm.tqdm(range(2,10)):
    clf_MBK = MiniBatchKMeans(n_clusters=i)
    clf_MBK.fit(tfidf_reduced_20)
    score.append(silhouette_score(tfidf_reduced_20,clf_MBK.labels_,metric='euclidean'))
```

## Prepare data

```python
df_cat3.reset_index(drop=True,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_cat3['target']=le.fit_transform(df_cat3['target'])
```

```python
print(le.classes_)
```

```python
del df_cat3['Unnamed: 0']
```

```python
clf_df = df_cat3[['title_no_bracket','requirement','target']]
```

```python
text_train_clf =[]
for i in tqdm.tqdm(range(len(df_cat3))):
    text_train_clf.append(str(df_cat3['title_no_bracket'][i]+df_shuffle['requirement'][i]))
    
clf_text_clf = []
preprocess_text(text_train_clf,clf_text_clf)
```

```python
clf_df['feature']=clf_text_clf
```

```python
del clf_df['title_no_bracket']
```

```python
del clf_df['requirement']
```

```python
from sklearn.model_selection import train_test_split
X_train,y_train,X_test,y_test=train_test_split(clf_df['feature'],clf_df['target'],

                                               test_size=0.2,random_state=42)
```

```python
from time import time
from pprint import pprint
from sklearn.feature_extraction.text import TfidfVectorizer
import xgboost
from sklearn.decomposition import KernelPCA
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

estimators=[('Tfid',TfidfVectorizer()),
            ('KPCA',KernelPCA(n_components=100,kernel='rbf')),
            ('xgb',xgboost.XGBClassifier(use_label_encoder=False))]

pipeline = Pipeline(estimators)

parameters={
    'KPCA__n_components':(4,50,100,150),
    'xgb__max_depth':(8,7,6,5),
    'xgb__n_estimators':(100,300,500)
}

grid_search = GridSearchCV(pipeline,parameters,n_jobs=-1,verbose=1)
print("Performing grid search...")
print("pipeline:", [name for name, _ in pipeline.steps])
print("parameters:")
pprint(parameters)

t0 = time()

grid_search.fit(X_train, X_test)

print("done in %0.3fs" % (time() - t0))
print()

print("Best score: %0.3f" % grid_search.best_score_)
print("Best parameters set:")
best_parameters = grid_search.best_estimator_.get_params()
for param_name in sorted(parameters.keys()):
    print("\t%s: %r" % (param_name, best_parameters[param_name]))
```

```python
estimators=[('Tfid',TfidfVectorizer()),
            ('KPCA',KernelPCA(n_components=100,kernel='rbf')),
            ('xgb',xgboost.XGBClassifier(use_label_encoder=False,
                                        n_estimators=300))]
model_clf = Pipeline(estimators)
model_clf.fit(X_train,X_test)
```

## Model validation

### Confusion matrix

```python
from sklearn.metrics import *
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(model_clf,y_train,y_test,cv=3)
```

```python
print('score: ',accuracy_score(y_test,y_train_pred))
conf_mx=confusion_matrix(y_test,y_train_pred)
print('confusion_matrix: \n',conf_mx)
plt.matshow(conf_mx,cmap='Blues_r')
```

### Classification_report

```python
target_names = ['Finance','IT','Others']
clf_report=classification_report(y_test,y_train_pred,target_names=target_names)
print(clf_report)
```
